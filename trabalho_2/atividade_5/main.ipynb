### Imports usados

```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

### Funções

```python
ALPHA = 0.0001
BETA = -0.08
LAMBDA = 5000
GAMA = 65

def f(q):
    return ALPHA*q**2 + BETA*q + LAMBDA/q + GAMA

def f_d1(q):
    return 2*ALPHA*q + BETA - LAMBDA/(q**2) 

def f_d2(q):
    return 2*ALPHA + 2*LAMBDA/(q**3)
```

### Método de newton

```python
def newton_method(q0):
    # init the values
    last_q  = q0
    q       = q0
    counter = 1
    arr     = []
    
    for _ in range(1000):
        # calculate the new value
        q = last_q - f_d1(q)/f_d2(q)

        # verify if the solution is close enough
        if abs(q - last_q) < 1e-6:
            break

        # update the last value
        last_q = q
        counter += 1
        if _ < 4:
            arr.append(q)

    return q, counter, arr

solution, counter, arr = newton_method(50)

print(f"Solution: {solution}")
print(f"Counter: {counter}")
print(f"Primeiras 4 iterações:\n{arr}")
```

### Função no intervalo [150, 800]

```python
x = [i for i in range(150, 800)]
y = [f(i) for i in x]

x0 = 500
y0 = f(x0)

plt.plot(x, y, color='green', label='f(q) = αq² + βq + λ/q + γ')
plt.scatter(x0, y0, color='red', label='[500, 60]')
plt.grid()
plt.legend()
plt.xlabel("q")
plt.ylabel("f(q)")
plt.show()

```
